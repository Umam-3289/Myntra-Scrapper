```python
import os
import time
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
from selenium import webdriver
product="THE BEAR HOUSE"
# define path to chrome webdriver executable

#Initialize the chrome webdriver
driver=webdriver.Chrome()

#Navigate the URL
driver.get("https://www.myntra.com/men-casual-shirts?rawQuery="+product)

#once the webpage is loaded you can extract the page text
myntra_text=driver.page_source

myntra_html=bs(myntra_text,"html.parser")
```

```python
myntra_html
```

```python
pclass=myntra_html.find_all("ul",{"class":"results-base"})
```

```python
pclass[0]
```

```python
for i in pclass:
    href=i.find_all('a',href=True)
    t=(href[0]['href'])

productLink="https://www.myntra.com/"+t
print(productLink)
```

```python
driver.get(productLink)
proRes=driver.page_source
proRes
```

```python
prores_html=bs(proRes,'html.parser')
prores_html
```

```python
Reviews=prores_html.find_all('a',{"class":"detailed-reviews-allReviews"})
Reviews
```

```python
t2=Reviews[0]['href']
review_link="https://www.myntra.com"+t2
review_link
```

```python
driver.get(review_link)
review_page=driver.page_source
```

```python
review_page
```

```python
review_html=bs(review_page,'html.parser')
review_html
```

```python
review=review_html.find_all("div",{"class":"detailed-reviews-userReviewsContainer"})
review
```

```python
# Fetching user_rating,user_name,user_comment from reviews
for i in review:
    user_rating=i.find_all('div',{"class":"user-review-main user-review-showRating"})
    user_comments=i.find_all('div',{"class":"user-review-reviewTextWrapper"})
    user_name=i.find_all('div',{"class":"user-review-left"})
```

```python
print(user_rating)
print(user_comments)
print(user_name)
```

```python
for i in range(len(user_rating)):
    rating=user_rating[i].find("span",{"class":"user-review-starRating"}).get_text().strip()
    comments=user_comments[i].text
    name=user_name[i].find('span').text
    date=user_name[i].find_all('span')[1].text
    print(f"Date:{date}\n Name:{name}\n Rating:{rating}\n Comments:{comments}\n")
```
